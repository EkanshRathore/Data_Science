```python
#4. Divide the data into Input and Output
#For x
x = df.iloc[:,2:4].values
print(x)
```

```python
#For y
y = df.iloc[:,4].values
print(y)
```

```python
#5. For Training and Testing variables -
#The data of x and y is split into training and testing here
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train ,y_test = train_test_split(x,y,random_state=0)
#By default, the training variable gets 75% and testing variable gets 25%
```

```python
print(x.shape)#100%
print(x_train.shape)#75%
print(x_test.shape)#25%
```

```python
print(y.shape)#100%
print(y_train.shape)#75%
print(y_test.shape)#25%
```

```python
#Normalisation/Scaling (Done only in Multivariate Datasets)
#Scaling is done only if we have different number of digits in the inputs or the columns we have taken like 2 and 5 digits
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
print(x_test)
print(x_train)
```

```python
#7. Apply a classifies/regressor/cluster
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#8. Fit the model 
#Fit the LogisticRegression Model(x_train,y_train)
model.fit(x_train,y_train)
```

```python
#9.Predicting the output
y_pred = model.predict(x_test)
print(y_pred)
```

```python
print(y_test)
```

```python
#10. Accuracy 
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```
