```python
#WEB SCRAPING = Extracting data from the web
#WE will be scraping reviews from amazon website  
#We are using Beautiful Soup = Tool For Web Scraping
#url= https://www.amazon.in/Apple-iPhone-256GB-Deep-Purple/product-reviews/B0BDK2FB2T/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews
```

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.amazon.in/Apple-iPhone-256GB-Deep-Purple/product-reviews/B0BDK2FB2T/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page = requests.get(url) # requesting the contents of the page from the above url
print(page)
# Response [200] is correct if you are getting Response [503]then there is an error if there is an error rerun the cell 7-9 times, if you still don't get the correct response, then check the url again.
```

```python
page.content #Prints the HTML source code of the entire page (Amazon customer Review page)
```

```python
soup = bs(page.content,'html.parser')
print(soup)
```

```python
#Now let us try to extract names of people
names = soup.find_all('span',class_ ='a-profile-name')[2:]
print(names)
```

```python
len(names)
```

```python
#Create  a Dataframe using pandas
import pandas as pd
df = pd.DataFrame(names,columns=['Customer Name'])
print(df)
```

```python
#Let us remove the duplicates
df.drop_duplicates(subset=None, keep="first", inplace=True)
print(df)
```

```python
len(df)
```

```python
#Let us scrape the review titles
titles = soup.find_all('a' ,class_ ='review-title')
print(titles)
```

```python
len(titles)
```

```python
#Data Cleaning and Data Filtering
review_title = []
 
for i in range(0,len(titles)):
    review_title.append(titles[i].get_text()[19:-1])
print(review_title)    
```

```python
#Review rating
#Review Dates
review_rating = soup.find_all('i',class_ ="review-rating")[2:]
print(review_rating)
```

```python
#Data Cleaning And Filtering 
review_rating1 = []

for i in range(0 , len(review_rating)) :
    review_rating1.append(titles[i].get_text()[0:12])
print(review_rating1)
```

```python
len(review_rating1)
```

```python
#Review Content
content = soup.find_all('span',class_ ='a-size-base review-text review-text-content')
print(content)
```

```python
len(content)
```

```python
#Data Cleaning and Filtering
review_content = []
 
for i in range(0 , len(content)):
     review_content.append(content[i].get_text()[1:-1])
print(review_content)
```

```python
#Review Dates 
#<span data-hook="review-date" class="a-size-base a-color-secondary review-date">Reviewed in India on 28 July 2023</span>
dates = soup.find_all('span' , class_ = 'a-size-base a-color-secondary review-date')[2:]
print(dates)
```

```python
len(dates)
```

```python
#Data Cleaning and Filtering
review_dates = []

for i in range (0 , len(dates)):
    review_dates.append(dates[i].get_text().replace("Reviewed in India on " , ''))
print(review_dates)    
```

```python
df['Review titles'] = review_title
df['Review Content'] = review_content
df['Review Rating'] = review_rating1
df['Review Date'] = review_dates
print(df)
```

```python
#Now let's turn the dataframe to csv file
df.to_csv('Amazon_1.csv')
```
